# Empirical bootstrap based confidental interval

```python
print(sk.utils.resample.__doc__)
dir(sk.utils)
```

```python
x=np.array([180,162,158,172,168,150,171,183,165,176])
#configure bootstrap
n_iteration=1000
n_size=int(len(x))
#run bootstrap
medians=[]
for i in range(n_iteration):
    #prepare trainand test size
    s=sk.utils.resample(x,n_samples=n_size)#UNIFORMLY CHOSING NUMBER FROM SAMPLE DATA FOR BOOTSTRAPING 
    m=np.median(s);
    medians.append(m)
#plot scores
pyp.hist(medians)#plot histogram
pyp.show()
#confidence intervals
alpha=0.95
p =((1.0-alpha)/2.0)*100
lower=np.percentile(medians,p)
p =(alpha+(1.0-alpha)/2.0)*100
upper=np.percentile(medians,p)
print(lower,upper)


```
